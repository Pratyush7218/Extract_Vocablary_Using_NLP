```python
import re
from nltk.tokenize import word_tokenize

def remove_special_int(text):
    # Tokenize the text into individual words
    tokens = word_tokenize(text)

    # Define a regex pattern to match special symbols
    pattern = r"[^a-zA-Z!@#$%^&*?'\s]"  # Keeps alphabets, digits, and whitespaces

    # Remove special symbols from each word
    cleaned_tokens = [re.sub(pattern, '', token) for token in tokens]

    # Join the cleaned tokens back into a paragraph
    cleaned_text = ' '.join(cleaned_tokens)

    return cleaned_text
```

```python
df = remove_special_int(text=df)
```

```python
def remove_symbal(text):
    # Tokenize the text into individual words
    tokens = word_tokenize(text)

    # Define a regex pattern to match special symbols
    pattern = r"[^a-zA-Z\s]"  # Keeps alphabets, digits, and whitespaces

    # Remove special symbols from each word
    cleaned_tokens = [re.sub(pattern, '', token) for token in tokens]

    # Join the cleaned tokens back into a paragraph
    cleaned_text = ' '.join(cleaned_tokens)

    return cleaned_text
```

```python
word = remove_symbal(text=df)
```

```python
stemmer = nltk.stem.porter.PorterStemmer()
```

```python
sword = nltk.corpus.stopwords.words('english')
```

```python
clean_words = [stemmer.stem(words) for words in word.split() if not words in sword]
```

```python
import csv
import xlwings
def append_list_to_csv(file_path, data_list):
    with open(file_path, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(data_list)
```

```python
import csv
from itertools import zip_longest

def transpose_csv_data(input_file_path, output_file_path):
    with open(input_file_path, 'r') as infile, open(output_file_path, 'w', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)

        transposed_data = zip_longest(*reader, fillvalue='')

        writer.writerows(transposed_data)

```

```python
output_file_path = 'Movie_Vocabulary.csv'
transpose_csv_data(input_file_path=file_path, output_file_path=output_file_path)
```
